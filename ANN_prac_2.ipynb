```python
#AND NOT MODEL
"""
*******************************************AND NOT MODEL******************************************************
"""
import pandas as pd
class Neuron:
    def __init__(self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def fire_nueron(self, inputs):
        weighted_sum = sum(w * x for w, x in zip(self.weights, inputs))# the summation function for the solving
        if weighted_sum <= self.threshold:
            return 0
        else:
            return 1

def ANDNOT(input1, input2):
    
    weights = [1, -1]
    threshold = 0      
    neuron = Neuron(weights, threshold)#object creation
    return neuron.fire_nueron([input1, input2])#activating the nueron 

# Create a DataFrame to store the results
results = {
    'X': [1, 1, 0, 0],
    'Y': [1, 0, 1, 0],
    'Output': [ANDNOT(1, 1), ANDNOT(1, 0), ANDNOT(0, 1), ANDNOT(0, 0)]
}

df = pd.DataFrame(results)#formig the table

# Display the DataFrame
print(df)
```
